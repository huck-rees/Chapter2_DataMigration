```python
import pandas as pd
import os
from pathlib import Path

def compile_channel_belt_areas(input_csv, output_csv):
    """
    Compile channel belt areas from individual river CSV files.
    
    Parameters:
    -----------
    input_csv : str
        Path to CSV containing river names and working directories
    output_csv : str
        Path where compiled results should be saved
    """
    
    # Read the input datasheet
    df_input = pd.read_csv(input_csv)
    
    # Initialize lists to store results
    river_names = []
    cb_areas_m2 = []
    cb_areas_km2 = []
    
    # Counter for tracking
    found_count = 0
    missing_count = 0
    
    # Iterate through each river in the input datasheet
    for idx, row in df_input.iterrows():
        river_name = row['river_name']
        working_dir = row['working_directory']
        
        # Construct the path to the channel belt CSV
        cb_csv_path = os.path.join(
            working_dir,
            'Zhao_etal_2025',
            'B',
            river_name,
            'output_annual',
            river_name,
            f'{river_name}_channel_belt.csv'
        )
        
        # Check if file exists
        if os.path.exists(cb_csv_path):
            try:
                # Read the channel belt CSV
                df_cb = pd.read_csv(cb_csv_path)
                
                # Check if multiple rows exist
                if len(df_cb) > 1:
                    print(f"WARNING: {river_name} has {len(df_cb)} rows. Using first row only.")
                
                # Extract cb_area_m2 from first row
                cb_area_m2 = df_cb['cb_area_m2'].iloc[0]
                
                # Calculate cb_area_km2
                cb_area_km2 = cb_area_m2 / 1e6
                
                # Append to results
                river_names.append(river_name)
                cb_areas_m2.append(cb_area_m2)
                cb_areas_km2.append(cb_area_km2)
                
                found_count += 1
                
            except Exception as e:
                print(f"ERROR reading {river_name}: {e}")
                
        else:
            missing_count += 1
    
    # Create output dataframe
    df_output = pd.DataFrame({
        'river_name': river_names,
        'cb_area_m2': cb_areas_m2,
        'cb_area_km2': cb_areas_km2
    })
    
    # Save to CSV
    df_output.to_csv(output_csv, index=False)
    
    # Print summary
    print(f"\n{'='*60}")
    print(f"Compilation complete!")
    print(f"{'='*60}")
    print(f"Files found and processed: {found_count}")
    print(f"Files missing/skipped: {missing_count}")
    print(f"Total rivers in input: {len(df_input)}")
    print(f"Output saved to: {output_csv}")
    print(f"{'='*60}\n")
    
    return df_output


# ==============================================================================
# USER INPUTS
# ==============================================================================

# Path to input CSV with river names and working directories
input_csv = r"E:\Dissertation\Data\GreenbergZhaoRees_river_datasheet.csv"

# Path where you want to save the compiled output
output_csv = r'E:\Dissertation\Data\Zhao_etal_2025\Zhao_compiled_channel_belt_areas.csv'

# ==============================================================================
# RUN
# ==============================================================================

df_results = compile_channel_belt_areas(input_csv, output_csv)

# Display results
print(df_results)
```
