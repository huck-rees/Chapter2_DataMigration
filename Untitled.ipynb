```python
"""
find_missing_main_channels.py

Scans the RivMapper Channels directory for each river listed in the master CSV
and identifies any reach/year combinations that are missing a main_channel.shp
file. Results are written to 'missing_mainchannel_line.csv' in the same
directory as the input CSV.

Author: James (Huck) Rees; PhD Student, UCSB Geography
"""

import os
import glob
import re
import pandas as pd


def find_missing_main_channels(csv_file: str):
    """
    Scan Channel folders for each RivMapper river in the master CSV and report
    any reach directories that contain no main_channel.shp files.

    Directory structure expected:
        {working_directory}/RiverMapping/Channels/{river_name}/reach_{ds_order}/{YYYY}/main_channel.shp

    Parameters:
        csv_file (str): Path to the master river datasheet CSV.

    Returns:
        pd.DataFrame: DataFrame of missing entries, also saved as
                      'missing_mainchannel_line.csv' alongside the input CSV.
    """
    river_data = pd.read_csv(csv_file)

    # Only RivMapper rivers use main_channel.shp files
    rivmapper_rivers = river_data[river_data['real_reach_length_method'] == 'RivMapper']
    print(f"Checking {len(rivmapper_rivers)} RivMapper river(s)...\n")

    missing_records = []

    for _, row in rivmapper_rivers.iterrows():
        river_name = row['river_name']
        working_directory = row['working_directory']

        channels_dir = os.path.join(
            working_directory, "RiverMapping", "Channels", river_name
        )

        # Check that the river's Channels directory exists at all
        if not os.path.isdir(channels_dir):
            print(f"  [MISSING DIR] {river_name}: Channels directory not found at {channels_dir}")
            missing_records.append({
                'river_name': river_name,
                'ds_order': None,
                'year': None,
                'issue': 'Channels directory not found',
                'path_checked': channels_dir
            })
            continue

        # Find all reach subdirectories
        reach_dirs = sorted(glob.glob(os.path.join(channels_dir, "reach_*")))

        if not reach_dirs:
            print(f"  [NO REACHES] {river_name}: No reach_* subdirectories found")
            missing_records.append({
                'river_name': river_name,
                'ds_order': None,
                'year': None,
                'issue': 'No reach subdirectories found',
                'path_checked': channels_dir
            })
            continue

        river_ok = True

        for reach_dir in reach_dirs:
            folder_name = os.path.basename(reach_dir)
            match = re.match(r"reach_(\d+)$", folder_name)
            if not match:
                continue
            ds_order = int(match.group(1))

            # Find all year subdirectories inside this reach
            year_dirs = sorted([
                d for d in glob.glob(os.path.join(reach_dir, "*"))
                if os.path.isdir(d)
            ])

            if not year_dirs:
                print(f"  [NO YEARS]   {river_name} / reach_{ds_order}: No year subdirectories found")
                missing_records.append({
                    'river_name': river_name,
                    'ds_order': ds_order,
                    'year': None,
                    'issue': 'No year subdirectories found',
                    'path_checked': reach_dir
                })
                river_ok = False
                continue

            for year_dir in year_dirs:
                year = os.path.basename(year_dir)
                shp_path = os.path.join(year_dir, "main_channel.shp")
                if not os.path.isfile(shp_path):
                    print(f"  [MISSING SHP] {river_name} / reach_{ds_order} / {year}: main_channel.shp not found")
                    missing_records.append({
                        'river_name': river_name,
                        'ds_order': ds_order,
                        'year': year,
                        'issue': 'main_channel.shp not found',
                        'path_checked': year_dir
                    })
                    river_ok = False

        if river_ok:
            print(f"  [OK]          {river_name}: all reaches have main_channel.shp files")

    # Build output DataFrame
    if missing_records:
        missing_df = pd.DataFrame(missing_records)
    else:
        missing_df = pd.DataFrame(columns=['river_name', 'ds_order', 'year', 'issue', 'path_checked'])
        print("\nNo missing main_channel.shp files found.")

    # Write output CSV alongside the input CSV
    output_path = os.path.join(os.path.dirname(csv_file), "missing_mainchannel_line.csv")
    missing_df.to_csv(output_path, index=False)
    print(f"\nResults saved to: {output_path}")
    print(f"Total missing entries: {len(missing_df)}")

    return missing_df


# ── Run ───────────────────────────────────────────────────────────────────────

csv_file_path = r"E:\Dissertation\Data\Greenbergetal2024_river_datasheet.csv"
missing_df = find_missing_main_channels(csv_file_path)
```
